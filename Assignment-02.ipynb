# 1. Python program to scrape data for “Data Analyst” Job position in “Bangalore” location

```python
# Import libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
# connect to the Webdriver
driver=webdriver.Chrome(r"C:\Users\Shipra Pachauri\Downloads\chromedriver_win32 (1)\chromedriver.exe")

# open naukri page on automated chrome browser
driver.get("https://www.naukri.com/")
```

```python
# entering designaton  
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')

# entering location
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Banglore')

# click on search
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
# Initializing the empty lists to hold our desired data
sunglass_brand_name=[]
sunglass_desc_name=[]
sunglass_price=[]
# Setting page number variables
start_page=0
end_page=3
# Loop to switch between pages
for page in range(start_page,end_page):
    # Fetching data of each element
    sunglasses_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    sunglass_desc_tags=driver.find_elements(By.XPATH,'//a[@class="IRpwTa" or @class="IRpwTa _2-ICcC"]')
    sunglass_price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    # On Page 3, making list of size 20 to make total of 100 
    if(page == 2):
        sunglasses_tags = sunglasses_tags[:20]
    # Creating list to make DataFram
    for i in range(len(sunglasses_tags)):
        sunglass_brand_name.append(sunglasses_tags[i].text)
        sunglass_desc_name.append(sunglass_desc_tags[i].text)
        sunglass_price.append(sunglass_price_tags[i].text)
    # Clicking on Next button to go to next page
    next_btn=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_btn.click()
    # Waiting to load the next page while skipping the page 3 as loop will break after that
    if(page < end_page - 1):
        time.sleep(10)

# Print total length of brand, description, and price list
print(len(sunglass_brand_name), len(sunglass_desc_name), len(sunglass_price))

# Making a DataFrame to hold all the lists
df_sunglass = pd.DataFrame({'Brand':sunglass_brand_name, 'Product Description':sunglass_desc_name, 'Price':sunglass_price})
df_sunglass
```

```python
# cpu filter as Intel core i7
cpu_filter=driver.find_element(By.XPATH,'//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/span')
cpu_filter.click()
```
