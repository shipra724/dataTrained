# 1 and 2. Python program to scrap Brand Name, Name of the Product, Price, Return/Exchange, Expected Delivery, Availability andProduct URL of guitar on amazon

```python
# open webdriver chrome
driver=webdriver.Chrome("chromedriver.exe")
# maximize the automated chrome window
driver.maximize_window()
time.sleep(3)
# open amazon on the the automated chrome window
driver.get("https://www.amazon.in/")
time.sleep(3)
```

```python
# taking input from user
var_input = input("Enter your item to search: ")
```

```python
# enter guitar on search bar
guitar_search_text=driver.find_element(By.XPATH,'//input[@class="nav-input nav-progressive-attribute"]')
guitar_search_text.send_keys(var_input)
# click search button
guitar_search_btn=driver.find_element(By.ID,'nav-search-submit-button')
guitar_search_btn.click()
```

```python
url_each_guitar=[]

for k in range(3):
    urls_guitar = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in urls_guitar:
        url_each_guitar.append(i.get_attribute("href"))
    if k<2 :
        next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        next_button.click()
        time.sleep(3)
print(len(url_each_guitar))
```

```python
brand_guitar = []
name_guitar = []
price_guitar = []
return_guitar = []
availability_guitar = []
expect_delivery_guitar = []

for i in url_each_guitar:
    driver.get(i)
    time.sleep(2)
    try:
        brand_guitar_element = driver.find_element(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        brand_guitar.append(brand_guitar_element.text)
    except NoSuchElementException:
        brand_guitar.append('-')
        
    try:
        name_guitar_element = driver.find_element(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        name_guitar.append(name_guitar_element.text)
    except NoSuchElementException:
        name_guitar.append('-')
        
    try:
        price_guitar_element = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        price_guitar.append(price_guitar_element.text)
    except NoSuchElementException:
        price_guitar.append('-')
    
    try:
        return_guitar_element = driver.find_element(By.XPATH, '//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        return_guitar.append(return_guitar_element.text)
    except NoSuchElementException:
        return_guitar.append('-')
    
    try:
        expect_delivery_guitar_element = driver.find_element(By.XPATH, '//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]')
        expect_delivery_guitar.append(expect_delivery_guitar_element.text)
    except NoSuchElementException:
        expect_delivery_guitar.append('-')
    
    try:
        availability_guitar_element = driver.find_element(By.XPATH, '//*[@id="availability"]/span')
        availability_guitar.append(availability_guitar_element.text)
    except NoSuchElementException:
        availability_guitar.append('-')
        
print(len(brand_guitar), len(name_guitar), len(price_guitar), len(return_guitar), len(availability_guitar), len(expect_delivery_guitar))
```

```python
#Creating data frame for Guitar Details
df_guitar_details = pd.DataFrame({'Brand':brand_guitar, 'Name':name_guitar, 'Price':price_guitar, 'Return Policy':return_guitar, 'Availability':availability_guitar, 'Expected Delivery':expect_delivery_guitar, 'Product URL':url_each_guitar})

#Closing the automated driver window
driver.close()

df_guitar_details
```

# 4. Python program to scrap Brand Name,, Colour, RAM, Storage(ROM), Camera, Display Size, Battery Capacity, Price, Product URL of smartphone on flipkart.

```python
# open automated chrome driver window
driver=webdriver.Chrome("chromedriver.exe")
# maximaze the automated chrome  window
driver.maximize_window()
# open flipkart on the automated chrome  window
driver.get("https://flipkart.com/")
```

```python
#Enter Oneplus Nord on the search bar
search_phone_name=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_phone_name.send_keys('Oneplus Nord')
# Click on the search icon 
search_phone_name_icon=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search_phone_name_icon.click()
```

```python
phone_name=[]
phone_colour=[]
phone_price=[]
phone_RAM=[]
phone_ROM=[]
phone_display=[]
phone_camera=[]
phone_battery=[]
product_url=[]

phone_name_tag=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
phone_price_tags=driver.find_elements(By.XPATH,'//div[@class="_25b18c"]')
phone_detail_tag=driver.find_elements(By.XPATH,'//div[@class="fMghEO"]')
product_url_tag = driver.find_elements(By.CLASS_NAME,"_1fQZEK")

for i in range(len(phone_name_tag)) :
    splitted_data_phone_name = phone_name_tag[i].text.split('(')
    splitted_data_phone_detail = phone_detail_tag[i].text.split('\n')
    splitted_RAM_ROM_detail = splitted_data_phone_detail[0].split('|')
    
    phone_name.append(splitted_data_phone_name[0]) # scrap Brand name
    phone_colour.append(splitted_data_phone_name[1].split(',')[0]) # scrap colour of the phone
    phone_RAM.append(splitted_RAM_ROM_detail[0].replace('RAM','').strip(' ')) # scrap RAM of the phone
    phone_ROM.append(splitted_RAM_ROM_detail[1].replace('ROM','').strip(' '))  # scrap ROM of the phone
    phone_display.append(splitted_data_phone_detail[1].replace(' Display',''))  # scrap dislay of the phone
    phone_camera.append(splitted_data_phone_detail[2])  # scrap camera of the phone
    phone_battery.append(splitted_data_phone_detail[3])  # scrap battery capacity of the phone
    phone_price.append(phone_price_tags[i].text)   # scrap price of the phone
    product_url.append(product_url_tag[i].get_attribute('href'))   # scrap product url of the phone

# print total length of phone_name,phone_colour,phone_price,phone_RAM,phone_ROM,phone_display,phone_camera,phone_battery,product_url
print(len(phone_name), len(phone_colour),len(phone_price),len(phone_RAM),len(phone_ROM), len(phone_display), len(phone_camera),len(phone_battery) ,len(product_url))

# make a dataframe
df_smartphone=pd.DataFrame({'Brand Name':phone_name, 'Colour':phone_colour, 'RAM':phone_RAM, 'ROM':phone_ROM, 'camera':phone_camera ,'Display size':phone_display, 'Battery capacity':phone_battery, 'Price':phone_price, 'Product url':product_url})

# csv file formation
df_smartphone.to_csv('OnePlusDetails.csv')
#Closing the automated web window
driver.close()

df_smartphone
```

```python
#Initializing all the list required
hostel_name=[]
hostel_distance=[]
hostel_ratings=[]
total_reviews=[]
overall_reviews=[]
privates_price=[]
dorm_price=[]
property_desc=[]
private_price_tag = []
dorm_price_tag = []

#Fetching web elements for each detail required
hostel_name_tag=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
hostel_distance_tag=driver.find_elements(By.XPATH,'//div[@class="subtitle body-3"]')[2:]
hostel_ratings_tag=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
total_reviews_tag=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
overall_reviews_tag=driver.find_elements(By.XPATH,'//div[@class="keyword"]')[2:]
private_and_dorm_price_tag=driver.find_elements(By.XPATH,'//div[@class="price-col"]')
property_desc_tag = driver.find_elements(By.XPATH, '//div[@class="rating-factors prop-card-tablet rating-factors small"]')

#Loop to fill Private and Dorm price
for i in range(len(private_and_dorm_price_tag)):
    if i %2 == 0:
        private_price_text = private_and_dorm_price_tag[i].text.split('\n')[-1]
        privates_price.append(private_price_text)
    else :
        dorm_price_text_list = private_and_dorm_price_tag[i].text.split('\n')
        if len(dorm_price_text_list) > 1:
            dorm_price_text = dorm_price_text_list[-1].split(' ')[-1]
        else :
            dorm_price_text = dorm_price_text_list[0]
        dorm_price.append(private_price_text)
#For Loop to fill remaining details for the hostel
for i in range(len(hostel_name_tag)):
    hostel_name.append(hostel_name_tag[i].text)
    hostel_distance.append(hostel_distance_tag[i].text)
    hostel_ratings.append(hostel_ratings_tag[i].text)
    total_reviews.append(total_reviews_tag[i].text.split(' ')[0])
    overall_reviews.append(overall_reviews_tag[i].text)
    property_desc.append(property_desc_tag[i].text.replace('\n',','))
#Printing the hostel details
print(len(hostel_name),len(hostel_distance),len(hostel_ratings),len(total_reviews),len(overall_reviews),len(privates_price),len(dorm_price),len(property_desc))
# Making a dataframe
df_hostels=pd.DataFrame({'Hostel Name':hostel_name, 'Distance from city centre':hostel_distance ,'Ratings':hostel_ratings, 'Total Reviews':total_reviews, 'Overall reviews':overall_reviews, 'Privates from price':privates_price,'Dorms from price':dorm_price,'Facilities and propert description':property_desc})
df_hostels
```
