# 2 . python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)

```python
link=requests.get("https://www.imdb.com/chart/top/")
movies=BeautifulSoup(link.content)

movie_names=[] 
year_movie = []
movie_ratings=[]

movie_year_name_list = movies.find_all('td',"titleColumn")
movie_ratings_list = movies.find_all('td',"ratingColumn imdbRating")

for i in range(100):
    movie_name_year = movie_year_name_list[i].text.strip('\n').replace('\n','').split('(')
    movie_names.append(movie_name_year[0])
    year_movie.append(movie_name_year[1].replace(')',''))
    movie_ratings.append(movie_ratings_list[i].text.strip('\n'))

df_movies=pd.DataFrame({'Name':movie_names, 'Year':year_movie, 'Rating':movie_ratings})
df_movies
```

# 4. python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
pres_url=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
pres_source=BeautifulSoup(pres_url.content)

name = []
term_of_office = []
for i in pres_source.find_all("div","presidentListing"):
    president_details = i.text.split("\n")
    name.append(president_details[1].split("(")[0].strip(" "))
    term_of_office.append(president_details[2].split(":")[1].strip(" "))

df_president = pd.DataFrame({"President Name":name, "Term of Office":term_of_office})
df_president
```

# 5(a). python program to scrape Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. from icc-cricket.com- 

```python
url_odi=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page_odi=BeautifulSoup(url_odi.content)

odi_teams_list=page_odi.find_all("span","u-hide-phablet")
odi_details_list=page_odi.find_all("td","table-body__cell u-center-text")
odi_ratings_list=page_odi.find_all("td","table-body__cell u-text-right rating")

men_teams=[]
odi_matches=[]
odi_points=[]
odi_ratings=[]

odi_ratings.append(page_odi.find("td","rankings-block__banner--rating u-text-right").text.replace("\n","").strip(" "))
odi_matches.append(page_odi.find("td","rankings-block__banner--matches").text.replace("\n","").strip(" "))
odi_points.append(page_odi.find("td","rankings-block__banner--points").text.replace("\n","").strip(" "))
men_teams.append(odi_teams_list[0].text)

for i in range(9) :
    men_teams.append(odi_teams_list[i+1].text)
    odi_ratings.append(odi_ratings_list[i].text)
    odi_matches.append(odi_details_list[2*i].text)
    odi_points.append(odi_details_list[2*i +1].text)

df_odidetails=pd.DataFrame({'Teams':men_teams,'Matches':odi_matches,'Points':odi_points,'Ratings':odi_ratings})
df_odidetails
```

# 7. Python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world 

```python
url_cnbc = requests.get("https://www.cnbc.com/world/?region=world")
page_cnbc = BeautifulSoup(url_cnbc.content)

cnbc_headline=[]
cnbc_time=[]
cnbc_newslink=[]

cnbc_time_list = page_cnbc.find_all("time","LatestNews-timestamp")
cnbc_headline_list = page_cnbc.find_all("a","LatestNews-headline")

for i in range(len(cnbc_time_list)) :
    cnbc_time.append(cnbc_time_list[i].text)
    cnbc_headline.append(cnbc_headline_list[i].text)
    cnbc_newslink.append(cnbc_headline_list[i].get('href'))
    
df_cnbc = pd.DataFrame( { "Headline":cnbc_headline, "Time": cnbc_time, "News Link":cnbc_newslink} )
df_cnbc
```

# 10. python program to scrape the details of top publications from Google Scholar from(Rank,Publication,h5-index,h5-median)

```python
scrap_pub=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
formdetails=BeautifulSoup(scrap_pub.content)

rank=[]
pub=[]
index=[]
median=[]

for i in formdetails.find_all("td","gsc_mvt_t"):
    pub.append(i.text)

for i in formdetails.find_all("td","gsc_mvt_p"):
    rank.append(i.text)

deatils_list=formdetails.find_all("td","gsc_mvt_n")
for i in range(len(deatils_list)):
    if i %2==0:
        index.append(deatils_list[i].text)
    else:
        median.append(deatils_list[i].text)

df_details=pd.DataFrame({'Rank':rank, 'Publication':pub, 'hf_index':index,'hf_median':median})
df_details
```
