# 1.Python program to scrape the details(Rank,name,artist,upload date,views) of most viewed videos on YouTube from Wikipedia.

```python
# open webdriver chrome
driver=webdriver.Chrome("chromedriver.exe")
# maximize the automated chrome window
driver.maximize_window()
# open amazon on the the automated chrome window
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
video_name=[]
video_rank=[]
artist_name=[]
video_views=[]
upload_date=[]
video_name_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')[:30]
video_rank_tags=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
artist_name_tags=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
video_views_tags=driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[4]')
upload_date_tags=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')

for i in range(len(video_name_tags)):
    video_name.append(video_name_tags[i].text)
    video_rank.append(video_rank_tags[i].text)
    artist_name.append(artist_name_tags[i].text)
    video_views.append(video_views_tags[i].text)
    upload_date.append(upload_date_tags[i].text)
    
# print total length    
print(len(video_rank),len(video_name),len(artist_name),len(video_views), len(upload_date))
# creating a dataframe
df_videos=pd.DataFrame({'Rank':video_rank, 'Name':video_name, 'Artist ': artist_name,'Views':video_views , 'Upload Date':upload_date})
df_videos
```

```python
# close the automated driver window
driver.close()
```

# 5. Python program to scrape the details of trending repositories on Github.com

```python
# open webdriver chrome
driver=webdriver.Chrome("chromedriver.exe")
# maximize the automated chrome window
driver.maximize_window()
# open amazon on the the automated chrome window
driver.get("https://github.com/")
```

```python
#click on open source
open_source_toggle=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
open_source_toggle.click()
#click on trending from dropdown
trending_link=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
trending_link.click()
```

```python
repository_title=[]
repository_desc=[]
contributors_count=[]
language_used=[]
url_contributors=[]
repository_title_tag=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]/a')
for i in repository_title_tag:
    repository_title.append(i.text)
    url_contributors.append(i.get_attribute('href'))
repository_desc_tag=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in repository_desc_tag:
    repository_desc.append(i.text)


    
```

```python
language_used_tag_div=driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]')
language_used=[]
language_used_tag=driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
count = 0
for i in range(len(language_used_tag_div)) :
    if language_used_tag_div[i].text[0:1].isdigit() == False :
        language_used.append(language_used_tag[count].text)
        count = count + 1
    else :
        language_used.append('')
```

```python
contributors_count=[]
for i in url_contributors :
    driver.get(i)
    time.sleep(1)
    contributors_count_tag = driver.find_elements(By.XPATH, '//span[@class="Counter"]')
    count = contributors_count_tag[len(contributors_count_tag) - 1].text
    if(len(count) > 0) :
        contributors_count.append(count)
    else :
        contributors_count.append(1) # single contributor, no list provided
```

```python

print(len(repository_title), len(repository_desc), len(contributors_count), len(language_used))
df_repository=pd.DataFrame({'Repository Title':repository_title, 'Repository Description':repository_desc, 'Contributors Count':contributors_count ,'Language':language_used })
df_repository
```

```python
#close automated driver window
driver.close()
```

```python
tv_shows_name=[]
tv_shows_year=[]
tv_shows_genre=[]
tv_shows_run_time=[]
tv_shows_ratings=[]
tv_shows_votes=[]
# scraping tv show name and tv show year
tv_shows_name_tags= driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
for i in range(100):
    tv_shows_name_list=tv_shows_name_tags[i].text.strip('\n').split('(')
    tv_shows_name.append(tv_shows_name_list[0])
    tv_shows_year.append(tv_shows_name_list[1])
# scraping tv show genre
tv_shows_genre_tags=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in tv_shows_genre_tags:
    tv_shows_genre.append(i.text)
# scraping tv show run time
tv_shows_run_time_tags=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in tv_shows_run_time_tags:
    tv_shows_run_time.append(i.text)
# scraping tv show rating
tv_shows_ratings_tags=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span[2]')
for i in tv_shows_ratings_tags:
    tv_shows_ratings.append(i.text)
# scraping tv show votes
tv_shows_votes_tags=driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"][3]/span[2]')
for i in tv_shows_votes_tags:
    tv_shows_votes.append(i.text)
    
# print total length 
print(len(tv_shows_name),len(tv_shows_year),len(tv_shows_genre),len(tv_shows_run_time),len(tv_shows_ratings),len(tv_shows_votes))
# creating a dataframe
df_tv_shows=pd.DataFrame({'TV Show Name':tv_shows_name , 'TV Show Year':tv_shows_year, 'TV Show Genre': tv_shows_genre,'TV Show Run Time': tv_shows_run_time,'TV Show Ratings':tv_shows_ratings , 'TV Show Votes':tv_shows_votes})
df_tv_shows

```

```python
data_set_name=[]
data_type=[]
default_task=[]
attribute_type=[]
instances=[]
attribute=[]
data_set_year=[]
data_set_name_tag = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table')
data_type_tag=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')[1:]
default_task_tag = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')[1:]   
attribute_type_tag = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')[1:]
instances_tag=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')[1:]
attribute_tag = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')[1:]
year_tag=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')[1:]
for i in range(len(data_set_name_tag)):
    data_set_name.append(data_set_name_tag[i].text)
    data_type.append(data_type_tag[i].text)
    default_task.append(default_task_tag[i].text)
    attribute_type.append(attribute_type_tag[i].text)
    instances.append(instances_tag[i].text)
    attribute.append(attribute_tag[i].text)
    data_set_year.append(year_tag[i].text)

print(len(data_set_name) , len(data_type) , len(default_task) , len(attribute_type), len(instances) , len(attribute) , len(data_set_year))

df_data_set=pd.DataFrame({'Dataset Name':data_set_name , 'Data Type':data_type , 'Task':default_task , 'Attribute types': attribute_type, 'no. of instances':instances, 'no. of attributes':attribute, 'Year':data_set_year})
df_data_set
```
